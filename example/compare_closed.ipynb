# impedance comparison for closed tube by original calcimp and calcimp-python

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

plt.style.use('seaborn-v0_8-darkgrid')
```

```python
import calcimp
```

```python
# read org calcimp result
ref = pd.read_csv("../sample/closed.imp")
```

```python
# calcimp-python result
frq, re, im, mag = calcimp.calcimp("../sample/closed.men")
```

```python
# plot comparison
plt.plot(frq, mag)
plt.plot(ref['freq'],ref['mag'] )
```

```python
# for ease of comparison, make a dataframe
res = pd.DataFrame({
    'freq': frq,
    're': re,
    'im': im,
    'mag': mag
})
```

```python
ref.head(10)
```

```python
res.head(10)
```

There is no difference between the two results. That means at least basic impedance calculation is correctly implemented in calcimp-python.

But it is known that for open end tube, the result is not consitent.
That means the end radiation effect is different. And that might come from cephes functions.
