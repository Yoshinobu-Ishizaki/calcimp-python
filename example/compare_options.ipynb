## Calcimp 計算パラメータ違いによる結果比較 

```python
import calcimp
```

```python
import matplotlib.pyplot as plt
plt.style.use('seaborn-v0_8-darkgrid')

plt.rcParams["font.family"] = "Noto Sans CJK JP"
```

```python
myfile = "../sample/test.men"
```

```python
# calculate sample data with default options
f,re,im,mag = calcimp.calcimp(myfile)
```

```python
# without wall dumping
_,_,_, mag_nodump = calcimp.calcimp(myfile, dump_calc = False)
```

```python
fig, ax = plt.subplots()

plt.plot(f,mag, label="default")
plt.plot(f,mag_nodump, label="no_dumping")
ax.set_title("壁面摩擦効果比較")
ax.set_xlabel("Hz")
ax.set_ylabel("Mag(dB_SPL)")
ax.legend()
```

```python
# radiation effect BUFFLE
_,_,_, mag_buffle = calcimp.calcimp(myfile, rad_calc = calcimp.BUFFLE)
```

```python
# no radiation effect 
_,_,_, mag_none = calcimp.calcimp(myfile, rad_calc = calcimp.NONE)
```

```python
fig, ax = plt.subplots()

plt.plot(f,mag, label="default(PIPE)")
plt.plot(f,mag_buffle, label="BUFFLE")
plt.plot(f,mag_none, label="NONE")
ax.set_title("出口放射関数効果比較")
ax.set_xlabel("Hz")
ax.set_ylabel("Mag(dB_SPL)")
ax.legend()
```

```python
# high temperature
_,_,_,mag_hot = calcimp.calcimp(myfile, temperature = 30.0)
```

```python
fig, ax = plt.subplots()

plt.plot(f,mag, label="default(24C)")
plt.plot(f,mag_hot, label="30C")
ax.set_title("温度効果比較")
ax.set_xlabel("Hz")
ax.set_ylabel("Mag(dB_SPL)")
ax.legend()
```
